```python
import numpy as np
#%matplotlib notebook
import matplotlib.pyplot as plt
import os
```

---

```python
path = os.path.abspath("../data/intensity/atlas/QS")
data = {}
data["old"] = np.load(f"{path}/atlas_QS.npy.before20220107")
data["new"] = np.load(f"{path}/atlas_QS.20221118.npy")
data["old"].shape, data["new"].shape
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,5), dpi=100)
for key, val in data.items():
    fac  = 2.5 if key=="old" else 1.0
    wfac = 1 if key=="old" else 1E8
    style = "o-" if key=="old" else "-"
    ax.plot(val[0,:]*wfac, val[1,:]*fac, style, label=key, linewidth=1.0, alpha=0.5, markersize=1)
ax.set_xlabel("Wavelength [A]")
ax.set_ylabel("Intensity")
ax.legend()
ax.set_xlim(0,10000)
#ax.set_xlim(1200,1230)
#ax.set_xlim(6555,6570)
ax.set_xlim(8540,8550)
#ax.set_xlim(3925,3975)
ax.set_ylim(0, 5E14)
#ax.set_ylim(0, 2E13)
#ax.set_yscale("log")
plt.show()
```

```python
path = "D:\\share\\src\\projects\\20220620.spectra\\data\\intensity\\atlas\\QS"
data = np.load(f"{path}/atlas_QS.20221118.npy")
data.shape
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,6), dpi=150)

ax.plot(data[0,:]*1E8,data[1,:])
ax.set_ylabel("intensity [$erg/cm^{2}/sr/cm/s$]", fontsize=16)
ax.set_xlabel("wavepength [$\AA$]", fontsize=16)
ax.set_yscale("log")
ax.set_xscale("log")
print(ax.get_xticks())
ax.tick_params(axis='both', labelsize=16)
ax.set_xlim(1,1E7)
ax.set_ylim(1E3,1E15)
plt.show()
```
